```python
import matplotlib.pyplot as plt
import numpy as np

# Data
people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(age, height, label='Height')
plt.scatter(age, weight, label='Weight')
plt.xlabel('Age')
plt.ylabel('Height/Weight')
plt.title('Scatter Plot: Age vs Height/Weight')
plt.legend()
plt.show()

# Bar Chart
x = np.arange(len(people))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, height, width, label='Height')
rects2 = ax.bar(x + width/2, weight, width, label='Weight')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('People')
ax.set_ylabel('Height/Weight')
ax.set_title('Bar Chart: People vs Height/Weight')
ax.set_xticks(x)
ax.set_xticklabels(people)
ax.legend()

fig.tight_layout()
plt.show()

# Histogram
plt.hist(height, bins=5, alpha=0.5, label='Height')
plt.hist(weight, bins=5, alpha=0.5, label='Weight')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram: Height/Weight')
plt.legend(loc='upper right')
plt.show()
```
